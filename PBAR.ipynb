```python
from bs4 import BeautifulSoup as soup

from urllib.request import urlopen as uReq

```

```python
import pandas as pd

import json

import urllib

import re

from progressbar import ProgressBar

pbar = ProgressBar()
```

```python
d = []
df = pd.read_csv('NPITask - NPITask.csv')
df = df[:]
urls = df['Urls']
```

```python
for url in pbar(urls):
    try:
        client = uReq(url)
        page_html = client.read()
        #print(page_html)
        page_soup = soup(page_html)
        name = page_soup.find("h2",{"class":"g-font-weight-300 g-mr-10"}).text
        phone = page_soup.find("abbr",{"title":"Phone"}).findNext('a').text
        npi = page_soup.find("title",{"itemprop":"name"}).text
        image= page_soup.find('img',{"class":"img-fluid w-100"})['src']
        taxonomy = page_soup.find("a",{"class":"jumplink"}).text
        pattern = '(?<=\(NPI)(.+?)(?=\))'
        result = re.findall(pattern, npi)
        d.append(
                {"Name":name,
                 "PHONE":phone,
                 "NPI":result,
                 "IMAGE":f"https:{image}",
                 "TAXONOMY":taxonomy,
                 "urls":url})
    except Exception:
        pass
 
```

```python
the_data = pd.DataFrame(d)
```
